Common Options for newline=
OptionÔæ†Description
newline=NoneÔæ†Default. Universal newline mode: all \n, \r\n, and \r are translated to \n on reading and back to system default on writing.
newline=''Ôæ†No translation: newline characters are written and read as-is. Often used with csv to prevent extra blank lines on Windows.
newline='\n'Ôæ†Use only line feed. Suitable for Unix/Linux-style newlines.
newline='\r\n'Ôæ†Use carriage return + line feed. Standard for Windows text files.
newline='\r'Ôæ†Use carriage return only (classic Mac-style, rarely used today).


JSON

```python
import json
data={"d":34, "a":24, "b":35, "c":78 }
print(json.dumps(data, indent=2, sort_keys=True))

#json.dumps() is a function that converts a Python dictionary (data) into a JSON-formatted string.
#indent=2: Formats the output to be pretty-printed with an indentation of 2 spaces for readability.
```

```python
import json
data = {"name":"Frieda", "is_dog": True, "age":8}
data["color"] = "orange"
data["favorite_food"] = ["pizza", "ice cream"]
json_str = json.dumps(data)
print(json_str)


with open("output.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)

```

```python
# Step 1: Load JSON data from file
with open("output.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Step 2: Access specific values
print("Name:", data["name"])
print("Favorite food:", data["favorite_food"])
# print("Microchipped:", data.get("microchipped"))

# Step 3: Update some values
data["age"] += 2  # birthday!
data["favorite_food"] = "Burger"  # new favorite
data["last_vet_visit"] = "2025-07-20"  # updated visit

# Step 4: Add new entries
data["disliked_fool"] = "vegetables"
data["daily_study_hours"] = 4

# Step 5: Remove a key (if exists)
data.pop("likes_water", None)  # safely remove without error if not found

# Step 6: Loop through all key-value pairs
print("\nüê∂ Updated Dog Profile:")

for key, value in data.items():
    print(f"{key}: {value}")

# Optional: Save the updated data back to file
with open("output_updated.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)
```
